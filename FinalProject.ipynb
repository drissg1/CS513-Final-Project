### Twitter Handle Clean Up
Below I wrote a function that was mapped onto the TwitterHandle column that was previously created in OpenRefine
This did some further refining to isolate only the twitter handle in this column

```python
def twitter_function(handle):
    """
    Function used to map the twitter handles values
    to raw @ token, this done by stripping the 
    http address execept for the @handle
    """
    http_find = re.split("htts://twitter.com/",handle)
    https_find = re.split("https://twitter.com/",handle)
    www_find =  re.split("www.twitter.com/",handle)
    if len(http_find) > 1:
        return(http_find
               [-1])
    elif len(https_find) > 1:
        return (https_find[-1])
    elif len(www_find) > 1:
        return(www_find[-1])
    else:
        return handle

#Run the function
data['TwitterHandle'] = data['TwitterHandle'].map(twitter_function,na_action = 'ignore')
```
